```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path = '../GA_Labs/data/five_thirty_eight_data/unisex-names/unisex_names_table.csv'
df = pd.read_csv(path)
```

```python
# Examine the first 10 rows
df.head(10)
```

```python
# drop column 'Unnamed:0'
df = df.drop('Unnamed: 0', axis=1)
# Also could do del df['Unnamed: 0']
df.head(10)
```

```python
# Create a new DataFrsame containing only the first name
# and the share of male and female names
col_list = ['name', 'male_share', 'female_share']
df2 = df[col_list]
df2.head(10)
```

```python
# Select only rows 5-10
df2.loc[5:10]
```

```python
# Set name as the index
df2.set_index('name', inplace=True)
#could also have inplace=False and set equal to new df
df2.head()
```

```python
# is the index unique?
df2.index.is_unique
```

```python
# Create a new column called name_length with length of name
df2['name_length'] = df2.index.str.len()
df2.head()
```

```python
#select just names riley and casey and only
# male_share and female_share columns
# try using brackets and loc
names = ['Casey', 'Riley']
cols = ['male_share', 'female_share']
df2.loc[names, cols]
```

```python
# Select all names with a male_share < 40%
mask = df2.male_share < 0.4
masked_df = df2[mask]
masked_df.head()
```

```python
# try with .loc
df2.loc[df2.male_share < 0.4]
```

```python
# Create a new column called gender, which tells if name is
# on average more likely male or female
df2['gender'] = df2.male_share.apply(lambda x: 'male' if x >= 0.5 else 'female')
```

```python
df2.head()
```

```python
# mean and std of each column?
df2.describe()
```

```python
# what are the 3 longest and shortest names?
sorted_df = df2.sort('name_length')
sorted_df.head(3)
```

```python
sorted_df2 = df2.sort('name_length', ascending=False)
sorted_df2.head(3)
```

```python
# Could also do something like this:
# shortest_names = df2.sort_values('name_length).index[0:3].values
```

```python
# Do male or female names have longer means and medians?
df2.groupby('gender').name_length.mean()
```

```python
df2.groupby('gender').name_length.median()
```

```python
# could also do like this:
df2.groupby('gender').name_length.agg(['mean','median'])
```

```python
# plot length of name vs male_share
(df2.groupby('name_length')
    .male_share
    .mean()
    .plot(style='.k', figsize=(12,8)))
```

```python
# Which first letter has the highest percentage
# of primarily female names?
df2['first_letter'] = df2.index.str[0]
df2.head()
```

```python
gb = df2.groupby(['first_letter','gender']).female_share.count()
```

```python
gb_unstack = gb.unstack().fillna(0)
```
